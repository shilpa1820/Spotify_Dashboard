```python
import pandas as pd
```

```python
df= pd.read_csv("/content/artist_data.csv")
df.head()
```

```python
new_genres=df['genres'].astype('string')
new_genres
```

```python
new_genres=new_genres.str.replace('[','')
new_genres=new_genres.str.replace(']','')

new_genres=new_genres.str.replace("'",'')
new_genres

```

```python
df['genres']=new_genres
df
```

```python
df.to_csv('artist_data.csv')
```
